```python


# 3. Create Lookup Table: Median Age by Title AND Pclass
# This gives us a granular reference (e.g., "Rich Master" vs "Poor Master")
group_medians = df.groupby(['Title', 'Pclass'])['Age'].median()

print("Median Reference Table:")
print(group_medians)

# 4. Define Imputation Function
def impute_age_properly(row):
    # Only fill if Age is missing (NaN)
    if pd.isnull(row['Age']):
        # Look up the median for this passenger's Title and Pclass
        return group_medians.loc[row['Title'], row['Pclass']]
    return row['Age']

# 5. Apply Imputation
df['Age'] = df.apply(impute_age_properly, axis=1)

# 6. Final Clean: Rounding
# Round to nearest whole number, ensure min age is 1
df['Age'] = df['Age'].round()
df.loc[df['Age'] < 1, 'Age'] = 1.0

# 7. Verification Plot
plt.figure(figsize=(10, 6))
# Plot histogram to check distribution
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution After Proper Imputation')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# 1. Load the dataset
df = pd.read_csv('titanic_large.csv')

# ---------------------------------------------------------
# Feature Engineering: Create Family Name Column
# ---------------------------------------------------------
# The Name column format is "LastName, Title. FirstName"
# We split the string by comma (',') and take the first part (index 0)
df['FamilyName'] = df['Name'].str.split(',').str[0]

# Check the first few rows to verify
print("First 5 extracted Family Names:", df['FamilyName'].head(5).tolist())

# ---------------------------------------------------------
# Analysis: Family Fate (Best and Worst)
# ---------------------------------------------------------
# Step A: Count members in each family
family_counts = df['FamilyName'].value_counts()

# Step B: Filter for "Large" families (e.g., 5 or more members)
# This removes single travelers and small groups to focus on "Clans"
large_families_names = family_counts[family_counts >= 5].index

# Step C: Calculate survival rate for these specific families
large_family_survival = df[df['FamilyName'].isin(large_families_names)].groupby('FamilyName')['Survived'].mean().sort_values()

# Step D: Get the exact stats for Best and Worst
worst_family = large_family_survival.head(1)
best_family = large_family_survival.tail(1)

print("\n--- Family Fate Analysis ---")
print(f"Total number of large families (5+ members): {len(large_family_survival)}")
print(f"Worst Surviving Family: {worst_family.index[0]} with {worst_family.values[0]:.1%} survival")
print(f"Best Surviving Family:  {best_family.index[0]} with {best_family.values[0]:.1%} survival")
print(f"Spread (Difference): {(best_family.values[0] - worst_family.values[0])*100:.1f}%")

# ---------------------------------------------------------
# Visualization (Matplotlib only)
# ---------------------------------------------------------
# Let's visualize the Bottom 5 and Top 5 to see how "flat" the difference is
top_5 = large_family_survival.tail(5)
bottom_5 = large_family_survival.head(5)
plot_data = pd.concat([bottom_5, top_5])

plt.figure(figsize=(12, 6))

# Create a color list: Red for worst, Green for best
colors = ['firebrick']*5 + ['forestgreen']*5

plot_data.plot(kind='bar', color=colors, width=0.7)

plt.title('Survival Rates: Worst 5 vs Best 5 Large Families')
plt.ylabel('Survival Rate')
plt.xlabel('Family Name')

# Add a line for the global average survival rate
global_avg = df['Survived'].mean()
plt.axhline(y=global_avg, color='blue', linestyle='--', linewidth=2, label=f'Global Avg ({global_avg:.1%})')

plt.ylim(0, 0.6) # Set y-axis to a reasonable scale
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg

# 1. Load Data
df = pd.read_csv('titanic_large.csv')

# 2. Basic Cleaning
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Sex_Code'] = df['Sex'].map({'male': 0, 'female': 1})

# 3. Select Features
features = ['Pclass', 'Sex_Code', 'Age', 'SibSp', 'Parch', 'Fare']
X = df[features].values
y = df['Survived'].values

# Add Intercept (A column of 1s so the math works)
X = np.c_[np.ones(X.shape[0]), X]

# ---------------------------------------------------------
# THE MODEL: One Line of Code
# ---------------------------------------------------------
# Finds the weights 'w' that minimize the error squared
# Equation: w = (X^T * X)^-1 * X^T * y
w, residues, rank, s = linalg.lstsq(X, y)

# ---------------------------------------------------------
# Prediction
# ---------------------------------------------------------
# Calculate score: y = w0 + w1*x1 + w2*x2 ...
scores = X.dot(w)

# Threshold: If score > 0.5, predict 1 (Survived), else 0
predictions = (scores >= 0.5).astype(int)

# ---------------------------------------------------------
# Results
# ---------------------------------------------------------
accuracy = np.mean(predictions == y)
print(f"Simple Least Squares Accuracy: {accuracy:.2%}")

print("\nWeights:")
for name, val in zip(['Intercept'] + features, w):
    print(f"{name}: {val:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# 1. Load Data
df = pd.read_csv('titanic_large.csv')

# 2. Preprocessing & Binning (Same as before)
df['LogFare'] = np.log1p(df['Fare'])
bin_obj = pd.cut(df['LogFare'], bins=200)
fare_means = df.groupby(bin_obj, observed=False)['Survived'].mean()

# Extract X (Bin Centers) and Y (Probabilities)
x_val = np.array([i.mid for i in fare_means.index])
y_val = fare_means.values

# Drop empty bins
mask = ~np.isnan(y_val)
x_val = x_val[mask]
y_val = y_val[mask]

# 3. Run Regression
slope, intercept, r_val, _, _ = linregress(x_val, y_val)
y_pred = intercept + slope * x_val # The model's predictions

# 4. CALCULATE ACCURACY (ERROR METRICS)
# Residuals = Actual - Predicted
residuals = y_val - y_pred

mse = np.mean(residuals**2)       # Mean Squared Error
rmse = np.sqrt(mse)               # Root Mean Squared Error
mae = np.mean(np.abs(residuals))  # Mean Absolute Error

print("--- Model Accuracy Report ---")
print(f"R-Squared (Explained Variance): {r_val**2:.4f}")
print(f"RMSE (Average Error):           {rmse:.4f} (Avg prob error)")
print(f"MAE (Mean Absolute Error):      {mae:.4f}")

# 5. Visualizing the Error
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True, gridspec_kw={'height_ratios': [2, 1]})

# Top: The Fit
ax1.scatter(x_val, y_val, color='blue', alpha=0.5, label='Actual Data')
ax1.plot(x_val, y_pred, 'r--', lw=2, label=f'Model (RMSE={rmse:.2f})')
ax1.set_title('Regression Fit')
ax1.set_ylabel('Survival Probability')
ax1.legend()

# Bottom: The Errors (Residuals)
ax2.scatter(x_val, residuals, color='green', alpha=0.5)
ax2.axhline(0, color='black', ls='--')
ax2.set_title('Residuals (Where is the model wrong?)')
ax2.set_ylabel('Error Margin')
ax2.set_xlabel('Log(Fare)')

plt.tight_layout()
plt.savefig('regression_accuracy_check.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
# ---------------------------------------------------------
# EXPERIMENT: IMPACT OF BIN COUNT ON R-SQUARED
# ---------------------------------------------------------
bin_counts = [20, 50, 100, 200]
results = []

plt.figure(figsize=(10, 6))

for bins in bin_counts:
    # Binning
    bin_obj = pd.cut(df['LogFare'], bins=bins)
    fare_means = df.groupby(bin_obj, observed=False)['Survived'].mean()
    
    # Extract Data
    x_val = np.array([i.mid for i in fare_means.index])
    y_val = fare_means.values
    mask = ~np.isnan(y_val)
    x_val, y_val = x_val[mask], y_val[mask]
    
    # Regression
    slope, intercept, r_val, _, _ = linregress(x_val, y_val)
    results.append((bins, r_val**2))
    
    # Plotting (Only plotting the trend lines to avoid clutter)
    plt.plot(x_val, intercept + slope*x_val, label=f'{bins} Bins ($R^2$={r_val**2:.2f})')

# Add the raw scatter for context (using 50 bins for visibility)
# (Re-calculating 50 bins just for the scatter dots)
bin_obj = pd.cut(df['LogFare'], bins=50)
fare_means = df.groupby(bin_obj, observed=False)['Survived'].mean()
x_scat = np.array([i.mid for i in fare_means.index])
y_scat = fare_means.values
plt.scatter(x_scat, y_scat, color='black', alpha=0.1, label='Data Points (50 Bins)')

plt.title('The "Resolution Effect": How Bin Count Changes Correlation Strength')
plt.xlabel('Log(Fare)')
plt.ylabel('Survival Probability')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('bin_resolution_effect.png')
plt.show()

print("\n--- Bin Count Impact ---")
for r in results:
    print(f"Bins: {r[0]:<4} -> R-Squared: {r[1]:.4f}")
```
