# Checking Duplicates

```python
# Function to calculate the number of duplicate values each column
def check_duplicates(df):
    duplicates = {}
    for col in df.columns:
        duplicates[col] = df[col].duplicated().sum()
    return pd.Series(duplicates)

# Applying the Function
result = check_duplicates(df)
print(result)
```

Since we are dealing with a dataset where approximately 20% of the entries for 'Age' are missing, simply removing all affected rows is often not a viable strategy.

Mean, Median, or Mode imputation would crowd data around single points, introducing bias. Instead, a sophisticated title-based imputation strategy was implemented:
- Like extracting the Title from name and cleaning the Data

```python
# Extract titles from 'Name'
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.')

# Define a function to assign a random age based on the extracted title
def assign_age(title):
    if title == 'Master' or title=="Miss":
        return np.random.randint(1, 19)   # Age range for 'Master' or 'Miss'
    else:
        return np.random.randint(18, 86)  # Age range for all the other titles

# Fill only the missing 'Age' values
df.loc[df['Age'].isnull(), 'Age'] = df.loc[df['Age'].isnull(), 'Title'].apply(assign_age)
```

```python
# Finding 'bad data' entries where age and title are inconsistent
bad_data = (
    ((df['Title'].isin(['Master', 'Miss'])) & (df['Age'] > 18)) |
    ((df['Title'].isin(['Mr', 'Mrs', 'Rev', 'Dr', 'Ms'])) & (df['Age'] < 18))
)

# Total length
length = len(df)

# Filtering the bad data rows
df = df[~bad_data]

# Display before and after
print("Rows removed:", length - len(df))
print("Final clean rows:", len(df))
```

```python
# Check if there is no Null Values
df.isnull().sum()
```

```python
# Define a function to clean the 'Age' column by ensuring ages are at least 1 and rounded to the nearest integer
def clean_age(age):
    return 1 if age < 1 else round(age)

# Apply the `clean_age` function to the 'Age' column of the DataFrame
df['Age'] = df['Age'].apply(clean_age)
```

```python
# For the number of unique values per column
def check_unique_values(df):
    unique = {}
    for col in df.columns:
        unique[col] = len(df[col].unique())
    return pd.Series(unique)

# Apply the function
result = check_unique_values(df)
print(result)
```

```python
# Create age bins in steps of 10 years
age_bins = pd.cut(df['Age'], bins=range(0, 91, 10), right=False)

# Group the DataFrame by the created Age bins and 'Survived' status
age_survival = df.groupby([age_bins, 'Survived']).size().unstack(fill_value=0)

# Rename the columns Dead[0] and Alive[1]
age_survival.columns = ['Dead', 'Alive']

# Plot a side-by-side bar chart
ax = age_survival.plot(kind='bar', figsize=(12, 6), width=0.8)

plt.title('Survival Status by Age (10-year bins)')
plt.xlabel('Age Range')
plt.ylabel('Number of Passengers')
plt.legend(title='Survival Status')
plt.xticks(rotation=45)

# Number to display on top
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2,
                height + 1,
                f"{height}", ha='center', va='bottom', fontsize=9)

plt.tight_layout() # Adjust plot layout to ensure all elements are visible
plt.show()
```

## Feature Engineering

```python
# Extract titles from the 'Name' column
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.')

# Calculate `Family_size` by summing 'SibSp' (siblings/spouses) and 'Parch' (parents/children) and adding 1 for the passenger itself
df['Family_size'] = (df['SibSp'] + df['Parch'] + 1)
```

```python
# Define a function to categorize passengers into 'Child', 'Adult', or 'Senior' age groups
def age_group(age):
    if age <= 18:
        return 'Child'
    elif 18 < age <= 60:
        return 'Adult'
    else:
        return 'Senior'

# Apply the `age_group` function to the 'Age' column to create a new 'AgeGroup' feature
df['AgeGroup'] = df['Age'].apply(age_group)
```

```python
# Define a function to categorize ticket fares into 'High Fare', 'Low Fare', or 'Medium'
def categorize_fare(fare):
    if fare > 50:
        return 'High Fare (>$50)'
    elif fare < 10:
        return 'Low Fare (<$10)'
    else:
        return 'Medium'

# Apply the `categorize_fare` function to the 'Fare' column to create a new 'FareGroup' feature
df['FareGroup'] = df['Fare'].apply(categorize_fare)
```

```python
# Create a mapping dictionary for 'Embarked' codes to their full port names
port_mapping = {'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}

# Map the 'Embarked' column using the dictionary to create a new 'Port_Name' column with descriptive names
df['Port_Name'] = df['Embarked'].map(port_mapping)
```

#### Statistical Test (Chi-Square)

```python
# Create contingency table of AgeGroup vs Survived
contingency = pd.crosstab(df['AgeGroup'], df['Survived'])

# Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency)

# Plot stacked bar chart of survival counts by AgeGroup
ax = contingency.plot(kind='bar', stacked=True, figsize=(10,6), color=['red','green'])

plt.title(f'Survival Counts by Age Group\nChi-Square p-value = {p:.2e}')
plt.xlabel('Age Group')
plt.ylabel('Number of Passengers')
plt.legend(title='Survival Status', labels=['Dead','Alive'])
plt.xticks(rotation=0)

# Annotate counts on bars
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2,
                    bar.get_y() + height/2,
                    f"{int(height)}",
                    ha='center', va='center', fontsize=9, color='white')

plt.tight_layout()
plt.show()
```

```python

# Display header for statistical results and researched reasons
print("\n--- STATISTICAL RESULTS & RESEARCHED REASONS ---")

# --- Q1 Analysis: The Wealth Gap (Statistical Test) ---
# Extract 'Survived' status for 'High Fare' and 'Low Fare' groups
high_fare = df[df['FareGroup'] == 'High Fare (>$50)']['Survived']
low_fare = df[df['FareGroup'] == 'Low Fare (<$10)']['Survived']

# Perform an Independent Samples T-test to compare survival means between high and low fare groups
t_stat, p1 = ttest_ind(high_fare, low_fare)

print(f"\n[Q1 Analysis: The Wealth Gap]")
print(f"P-Value: {p1}")

# Interpret the p-value for statistical significance
if p1 < 0.05:
    print("CONCLUSION: SIGNIFICANT.")
    print("Passengers paying High Fares survived at a statistically higher rate.")
    print("RESEARCHED REASON: First-class cabins (High Fare) were located on higher decks, closer to the lifeboats.")
    print("Additionally, locked gates often prevented 3rd class passengers from reaching the boat deck in time.")
else:
    print("CONCLUSION: NOT SIGNIFICANT.")
    print("Wealth did not affect survival chances.")

# --- Q2 Analysis: The Gender Gap (Statistical Test) ---
# Extract 'Survived' status for 'female' and 'male' passengers
female = df[df['Sex'] == 'female']['Survived']
male = df[df['Sex'] == 'male']['Survived']

# Perform an Independent Samples T-test to compare survival means between genders
t_stat, p2 = ttest_ind(female, male)

print(f"\n[Q2 Analysis: The Gender Gap]")
print(f"P-Value: {p2}")

# Interpret the p-value for statistical significance
if p2 < 0.05:
    print("CONCLUSION: SIGNIFICANT.")
    print("Females survived at a drastically higher rate than Males.")
    print("RESEARCHED REASON: The 'Birkenhead Drill' protocol ('Women and Children First') was strictly enforced by officers.")
    print("Men were often physically blocked from entering lifeboats at gunpoint to prioritize women.")
else:
    print("CONCLUSION: NOT SIGNIFICANT.")
    print("Gender did not affect survival chances.")
```

```python
# 4. STATISTICAL TEST (T-Test) for Embarkation Port
print("\n--- STATISTICAL RESULTS ---")

# Hypothesis: Passengers from Cherbourg (C) survived more than Southampton (S).
# We isolate the 'Survived' data for these two specific embarkation ports.
cherbourg_survived = df[df['Embarked'] == 'C']['Survived']
southampton_survived = df[df['Embarked'] == 'S']['Survived']

# Perform an Independent Samples T-Test to compare the survival rates
t_stat, p_value = ttest_ind(cherbourg_survived, southampton_survived)

print(f"Comparison: Cherbourg (C) vs Southampton (S)")
print(f"P-Value: {p_value}")

# Provide a conclusion based on the p-value
if p_value < 0.05:
    print("\nCONCLUSION: SIGNIFICANT.")
    print("Passengers boarding at Cherbourg had a statistically higher survival rate.")
    print("Possible Reason: Cherbourg likely had a higher percentage of 1st Class passengers.")
else:
    print("\nCONCLUSION: NOT SIGNIFICANT.")
    print("The port of embarkation did not affect survival chances.")
    print(f"The P-Value of {p_value:.3f} indicates that the survival difference is purely random.")
    print("There is zero statistical evidence that boarding at one port was safer than the other.")
```

## Machine Learning:

```python
# FARE ELASTICITY (Linear Regression)

# Apply log1p such that the zero fare will not raise error
df['LogFare'] = np.log1p(df['Fare'])

# Bin the 'LogFare' into 20 equal-width bins.
bin_obj = pd.cut(df['LogFare'], bins=20)

# Group by these bins and calculate the mean survival rate within each bin
fare_means = df.groupby(bin_obj)['Survived'].mean()

x_val = np.array([interval.mid for interval in fare_means.index])
y_val = fare_means.values

# Remove bins where survival probability is NaN
mask = ~np.isnan(y_val)
x_val = x_val[mask]
y_val = y_val[mask]

# Perform linear regression to find the relationship between LogFare and Survival Probability
slope, intercept, r_val, p_val, std_err = linregress(x_val, y_val)

print("\n--- Q3: Price Elasticity of Survival ---")
print(f"Slope: {slope:.4f} (Probability increase per Log")
print(f"R-Squared: {r_val**2:.4f} (Correlation Strength)")

# Visualization of Fare Elasticity
plt.figure(figsize=(10, 6))
plt.scatter(x_val, y_val, color='blue', label='Actual Survival (Binned)')
# Plot the linear regression line
plt.plot(x_val, intercept + slope*x_val, 'r-', linewidth=2, label=f'Trend (R2={r_val**2:.2f})')
plt.title(f"Fare Elasticity: How Money Bought Life\n(Slope={slope:.3f})")
plt.xlabel("Log(Fare)")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True, alpha=0.3) # Add grid
plt.show()
```

The analysis proves the "Pay-to-Survive" reality on the Titanic.

- There is a direct, mathematical link between ticket price and life. As the fare went up, the chance of survival went up linearly.

- The Numbers:
  - Cheap Tickets: Passengers paying low fares had a less than 25% of surviving.

  - Expensive Tickets: Passengers paying high fares had greater than 60% of surviving.

Wealth plays an important role in Suvival in Titanic as the R2 value closer to 1 the more reliable that money play's the vital role and the R value is 0.83(Which is closer to 1).
