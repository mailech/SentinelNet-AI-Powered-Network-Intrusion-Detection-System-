```python
import pandas as pd
import numpy as np

df = pd.read_csv('labeled_dataset.csv')

skewed_cols = ['duration', 'src_bytes', 'dst_bytes']

# Calculate the correlation coefficients
corr_duration_src = df['duration'].corr(df['src_bytes'])
corr_duration_dst = df['duration'].corr(df['dst_bytes'])

print("--- Correlation of 'duration' with other skewed columns ---")
print(f"duration and src_bytes : {corr_duration_src:.4f}")
print(f"duration and dst_bytes : {corr_duration_dst:.4f}\n")

df_log_transformed = pd.DataFrame()

for col in skewed_cols:
    # If log 0 infinity so we are adding 1
    df_log_transformed[f'{col}_log'] = np.log1p(df[col])

# 3. Save Log Values to a Separate File
output_filename = 'log_transformed_features.csv'
df_log_transformed.to_csv(output_filename, index=False)

print(f"Success! Log-transformed values saved to '{output_filename}'")
```
