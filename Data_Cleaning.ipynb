```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure the visual style for seaborn
sns.set_theme(style="whitegrid")

# Load your dataset
df = pd.read_csv('labeled_dataset.csv')

# --- 1. Target Variable Analysis ---
# Get the top 10 most frequent classes
top_10_classes = df['class'].value_counts().iloc[:]

plt.figure(figsize=(10, 6))
# Using seaborn countplot
sns.countplot(data=df, y='class', order=top_10_classes.index, palette='viridis')
plt.title('Top 10 Traffic Classifications (Normal vs Attack Types)')
plt.xlabel('Number of Network Packets')
plt.ylabel('Traffic Classification')
plt.tight_layout()
plt.show()

# --- 2. Categorical Feature Analysis (Protocol Types) ---
plt.figure(figsize=(8, 8))
protocol_counts = df['protocol_type'].value_counts()

# Using matplotlib pie chart
plt.pie(protocol_counts, labels=protocol_counts.index, autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Network Protocols')
plt.show()

# --- 3. Correlation Heatmap ---
# Selecting a subset of logical features to avoid a cluttered graph
subset_cols = ['src_bytes', 'dst_bytes', 'logged_in', 'count', 'srv_count', 
               'serror_rate', 'rerror_rate', 'same_srv_rate', 'diff_srv_rate']

corr_matrix = df[subset_cols].corr()

plt.figure(figsize=(10, 8))
# Using seaborn heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Selected Features')
plt.tight_layout()
plt.show()

# --- 4. Bivariate Analysis (Boxplot for Outliers/Distributions) ---
# Create a binary column: "Normal" or "Attack" for easier comparison
df['is_attack'] = df['class'].apply(lambda x: 'Normal' if x == 'normal' else 'Attack')

plt.figure(figsize=(8, 6))
# Using seaborn boxplot
sns.boxplot(x='is_attack', y='count', data=df, palette='Set2')
plt.title('Connection Count vs Traffic Type')
plt.xlabel('Traffic Classification')
plt.ylabel('Number of Connections (count feature)')
plt.tight_layout()
plt.show()
```
