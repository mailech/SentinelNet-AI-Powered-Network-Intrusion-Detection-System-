```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual styling for the project
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the dataset
# Ensure 'labeled_dataset.csv' is in the same directory as your notebook
df = pd.read_csv('labeled_dataset.csv')

# Create a binary target column for anomaly vs normal detection
df['binary_target'] = df['class'].apply(lambda x: 'normal' if x == 'normal' else 'anomaly')

print(f"Dataset loaded successfully with {df.shape[0]} rows and {df.shape[1]} features.")
df.head()
```

```python
# Count the occurrences of normal vs anomaly traffic
target_counts = df['binary_target'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', 
        startangle=140, colors=['#4C72B0', '#C44E52'], explode=(0, 0.05))
plt.title('SentinelNet: Overall Traffic Distribution (Normal vs Anomaly)', fontsize=14)
plt.show()
```

```python
# Group data by protocol type and the binary target
protocol_dist = pd.crosstab(df['protocol_type'], df['binary_target'])

# Plot stacked bar chart
protocol_dist.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#C44E52', '#4C72B0'])
plt.title('Traffic Classification by Protocol Type', fontsize=14)
plt.xlabel('Protocol Type', fontsize=12)
plt.ylabel('Number of Connections', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Classification')
plt.tight_layout()
plt.show()
```

```python
# Select only numerical columns for correlation calculation
numeric_cols = df.select_dtypes(include=[np.number])

# Calculate correlation matrix
corr_matrix = numeric_cols.corr()

# Plot heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, fmt=".2f", vmin=-1, vmax=1)
plt.title('Feature Correlation Heatmap for Network Traffic', fontsize=16)
plt.tight_layout()
plt.show()
```
