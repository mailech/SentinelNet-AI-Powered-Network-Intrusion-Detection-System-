```python
import os
import requests
```

```python
DATA_DIR="data"
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)

URLs = {
    'KCDtrain.txt': 'https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTrain+.txt',
    'KCDtest.txt': 'https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTest+.txt'
}

def download_file(url, filename):
    local_path = os.path.join(DATA_DIR, filename)
    
    if os.path.exists(local_path):
        print(f"[*] {filename} already exists.")
        return
    
    print(f"[+] Downloading {filename}...")
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()
        
        with open(local_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"Success saved to {local_path}")
    except Exception as e:
        print(f"Failed to download {filename}: {e}")
```

```python
for filename, url in URLs.items():
    download_file(url, filename)
print(f"All files should now be in 'data' folder.")
```
